## Wielkości ekstensywne i intensywne

Ekstensywne - to takie które się bezpośrednio sumują gdy potraktujemy dwa oddzielne układy.
* Masa - dwie jednakowe bryłki ołowiu, razem - dwa razy cięższe.
* "Energia" (najlepiej cała wewnętrzna i jak pominiemy oddziaływania między układami). Dwa kawałki kiełbachy, $2x$ kilokalorii.
* Całkowity pęd układu
* Całkowity moment pędu układu
* Całkowita liczba cząstek
Intensywne - nie sumują się po zsumowaniu poszczególnych układów.
* Prędkość - gdy poszczególnymi układami są dwie poszczególne cząstki, zsumowanym układem nie jest układ o prędkości $2V$.
* Przyśpieszenie
* Prędkość kątowa
* Potencjał chemiczny
* Temperatura - patrz anegdota Feynmana o podręczniku matematyki dla podstawówki, który recenzował. Wprowadzenie do zadania sensowne - takie gwiazdy mają taką temperaturę, większe taką, mniejsze owaką, kolor się zmienia, itd. I nagle zonk, pytanie brzmiało: "Jaką będą miały temperaturę obie razem?"

## Proces termodynamiczny quasi-statyczny

Taki, który nie ma gwałtownych (takich, których nie możemy sympatycznie scałkować po $dV$ czy czym tam innym) zmian. Przechodzi w swojej ewolucji czasowej tylko przez stany równowagowe.

Proces odwracalny - taki, gdzie jak odwrócimy zmianę warunków zewnętrznych (zaczniemy chłodzić wodę zamiast grzać), wróci do stanu poprzedniego (zmiany fazy w izolowanym układzie, typu lód, zdają się być odwracalne).

Demon Maxwella - przechodzi przez stany równowagi (wypuszcza jedną cząstkę naraz przez bramkę), ale jest nieodwracalny (nie może zmusić cząstek które przepuścił do powrotu). Nie ma implikacji w stronę że jeśli quasi-statyczny to odwracalny! W drugą stronę już tak.

## Układy w termodynamice

* Zamknięty - nie wymienia cząstek z otoczeniem. Możemy go np. grzać, dostarczając energię. Czajnik. Naprawdę szczelny czajnik, ok?
* Izolowany - nie wymienia energii ani cząstek z otoczeniem. Korea Najlepsza.
* Otwarty - wymienia z otoczeniem co chce. Korea Południowa.

## Zasady termodynamiki

0. Dla każdego układu termodynamicznego w równowadze *(ważne! dla wzburzonego gazu nie istnieje jednoznaczenie określona temperatura!)* istnieje intensywna funkcja stanu $\theta$ - temperatura. Ma to-to relację przechodnią - jeśli $\theta_1 = \theta_2$ oraz $\theta_2 = \theta_3$, to $\theta_1=\theta_3$. Definiujemy ją tak, żeby energia (ciepło) zawsze płynęło samoistnie od ciała z większym $\theta$ do tego z mniejszym (inaczej ból głowy ze znakami w całkach). Można ją sobie przeskalować np. przez $+3$, odjąć $273 K$ byleby była monotoniczna w ten sam sposób

1. Zachowanie energii - zmiana energii układu jest równa sumie:

$\delta E = - p dV + v dP + \omega dL + \phi dq + T dS + ...$

W ogólności to jest taka suma:

$\delta E = \sum_i x_i dX_i$, gdzie $x_i$ to zmienna intensywna, a $X_i$ to sprzężona do niej zmienna ekstensywna. I jak spojrzeć powyżej to można ogarnąć że temperatura jest intensywna do sprzężonej, ekstensywnej entropii, która w takim razie powinna się sumować między układami... i faktycznie się sumuje.
