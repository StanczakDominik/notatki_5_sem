## Do meritum

```
struct elem
{
    int v;
    elem *next;
}
elem *head;
```

* printujemy wszystkie:

```
void printAll()
{
    elem *p = head;
    while (p)      // w C++11 można: p != nullptr
    {
        cout << p -> val;
        p = p->next;
    }
}
```
O(n) czasowo, O(1) pamięciowo. Innymi słowy, T: O(n), M: O(1).

* Przeglądamy wszystkie w tył, przez rekurencję:

```
void printRevAll(elem *head)
{
    if (!head) return;
    printRevAll(head->next);
    cout << head->val;
}
```
T: O(n), M: O(n)

* Szukamy pierwszego wystąpienia elementu, rozbudowujemy strukturę danych o pole `*sentinel`:
```
elem search(int v)
{
    sentinel -> val = v;
    elem *result = null;
    elem *p = head;
    while (p -> val != v)
    {
        p = p -> next;
    }
    return result;
}
```

* Insertujemy na początek:
```
void insert (int v)
{
    head = new list(v, head);
}
```

* Insertujemy na koniec:
```
void insertLast (int v)
{
    elem *last = new elem(v, nullptr);
    p = head;
    if (!p)
    {
        head = last;
        return;
    }
    while (p->next)
    {
        p = p -> next;
    }
    p -> next = last;
}
```
* Insertujemy w posortowaną listę, w odpowiednim miejscu:

```
void insertSort( int v )
{
    if (!head)
    {
        head = new elem(v, val);
        return;
    }
    if (head->val > v)
    {
        head = new elem(v, head);
        return;
    }
    elem *p = head;
    while (p->next->val < v)
    {
        p = p->next;
    }
    p -> next = new elem (v, p-> next);
}
```
